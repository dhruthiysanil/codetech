```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
```

```python
# Load the dataset
file_path = "D:\sample-data-10mins.xlsx" # Replace with the actual path to your file
data = pd.read_excel(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Summary statistics
summary_stats = data.describe()
print("Summary Statistics:")
print(summary_stats)

# Check for missing values
missing_values = data.isnull().sum()
print("\nMissing Values:")
print(missing_values)
```

```python
# Set the style of the visualizations
sns.set(style="whitegrid")

# Histograms for 'Amount' and 'Boxes Shipped'
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(data['Amount'], kde=True, bins=30)
plt.title('Distribution of Amount')

plt.subplot(1, 2, 2)
sns.histplot(data['Boxes Shipped'], kde=True, bins=30)
plt.title('Distribution of Boxes Shipped')

plt.tight_layout()
plt.show()
```

```python
# Scatter plot for Amount vs. Boxes Shipped
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Amount', y='Boxes Shipped', data=data)
plt.title('Amount vs. Boxes Shipped')
plt.xlabel('Amount')
plt.ylabel('Boxes Shipped')
plt.show()

# Correlation coefficient
correlation = data[['Amount', 'Boxes Shipped']].corr()
print("\nCorrelation Coefficient:")
print(correlation)
```
